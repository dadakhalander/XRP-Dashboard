```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert Date column to datetime format
df1["Date"] = pd.to_datetime(df1["Date"])
```

```python
# Convert 'Change %' from string to float
df1["Change %"] = df1["Change %"].str.replace("%", "").astype(float) / 100
```

```python
# Reset index
df1 = df1.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot price trends
plt.figure(figsize=(12, 6))
plt.plot(df1["Date"], df1["Price"], label="XRP Price", color='blue')
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("XRP Price Trend Over Time")
plt.legend()
plt.show()
```

# avg price

```python
# Assuming df is already preprocessed and sorted by Date
df1["AvgPrice_5"] = df1["Price"].rolling(window=5).mean()
df1["AvgPrice_30"] = df1["Price"].rolling(window=30).mean()
df1["AvgPrice_365"] = df1["Price"].rolling(window=365).mean()
```

```python
df1.head(30)
```

```python
# Calculate ratios
df1["Ratio_5_30"] = df1["AvgPrice_5"] / df1["AvgPrice_30"]
df1["Ratio_5_365"] = df1["AvgPrice_5"] / df1["AvgPrice_365"]
df1["Ratio_30_365"] = df1["AvgPrice_30"] / df1["AvgPrice_365"]
```

```python
df1.tail(30)
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(14, 6))

# Plot price and moving averages
plt.plot(df1["Date"], df1["Price"], label="XRP Price", color="blue", alpha=0.6)
plt.plot(df1["Date"], df1["AvgPrice_5"], label="5-Day Avg", color="red", linestyle="dashed")
plt.plot(df1["Date"], df1["AvgPrice_30"], label="30-Day Avg", color="green", linestyle="dashed")
plt.plot(df1["Date"], df1["AvgPrice_365"], label="365-Day Avg", color="purple", linestyle="dashed")

plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("XRP Price & Moving Averages")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Plot ratios
plt.plot(df1["Date"], df1["Ratio_5_30"], label="Ratio 5/30", color="red")
plt.plot(df1["Date"], df1["Ratio_5_365"], label="Ratio 5/365", color="blue")
plt.plot(df1["Date"], df1["Ratio_30_365"], label="Ratio 30/365", color="green")

plt.axhline(y=1, color="black", linestyle="dashed")  # Reference line at 1
plt.xlabel("Date")
plt.ylabel("Ratio Value")
plt.title("XRP Moving Average Ratios")
plt.legend()
plt.show()
```

Insights from Visualization:

   If Ratio 5/30 > 1, short-term prices are rising compared to the 30-day trend (bullish signal).

   If Ratio 5/365 > 1, prices are rising above the yearly trend (strong bullish signal).

   If Ratio 30/365 < 1, the medium-term trend is lower than the yearly trend (bearish signal).

# Bollinger Bands

```python
# Calculate Bollinger Bands for AvgPrice_5, AvgPrice_30, AvgPrice_365
df1["Upper_Band_5"] = df1["AvgPrice_5"] + (2 * df1["StdPrice_5"])
df1["Lower_Band_5"] = df1["AvgPrice_5"] - (2 * df1["StdPrice_5"])

df1["Upper_Band_30"] = df1["AvgPrice_30"] + (2 * df1["StdPrice_30"])
df1["Lower_Band_30"] = df1["AvgPrice_30"] - (2 * df1["StdPrice_30"])

df1["Upper_Band_365"] = df1["AvgPrice_365"] + (2 * df1["StdPrice_365"])
df1["Lower_Band_365"] = df1["AvgPrice_365"] - (2 * df1["StdPrice_365"])
```

```python
# Plotting the Bollinger Bands
plt.figure(figsize=(14, 6))

# Plot for 5-day
plt.plot(df1["Date"], df1["AvgPrice_5"], label="5-Day Avg Price", color="blue")
plt.fill_between(df1["Date"], df1["Upper_Band_5"], df1["Lower_Band_5"], color="orange", alpha=0.3, label="Bollinger Bands 5-Day")

# Plot for 30-day
plt.plot(df1["Date"], df1["AvgPrice_30"], label="30-Day Avg Price", color="green")
plt.fill_between(df1["Date"], df1["Upper_Band_30"], df1["Lower_Band_30"], color="yellow", alpha=0.3, label="Bollinger Bands 30-Day")

# Plot for 365-day
plt.plot(df1["Date"], df1["AvgPrice_365"], label="365-Day Avg Price", color="purple")
plt.fill_between(df1["Date"], df1["Upper_Band_365"], df1["Lower_Band_365"], color="pink", alpha=0.3, label="Bollinger Bands 365-Day")

plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("XRP Bollinger Bands with Moving Averages")
plt.legend()
plt.show()
```

Insights from the Bollinger Bands Plot:

When the price moves close to the upper band, it suggests the asset is overbought (high volatility).

When the price moves close to the lower band, it suggests the asset is oversold (high volatility).

Narrow bands indicate low volatility, while wide bands signal higher volatility.
